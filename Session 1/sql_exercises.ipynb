## 1. Import libraries and set-up SQL engine

```python
from sqlalchemy import create_engine, text

import pandas as pd


DATABASE_PATH = "database/student.db"
ENGINE = create_engine(f"sqlite:///{DATABASE_PATH}")

def sql(query: str, engine) -> pd.DataFrame:
    with engine.connect() as conn:
        return pd.read_sql_query(text(query), conn)
```

## 3. Apply SQL commands from database

```python
query = "SELECT * FROM STUDENT_NAMES;"
sql(query, engine=ENGINE)
```

```python
query = "SELECT * FROM STUDENT_GRADES;"
sql(query, engine=ENGINE)
```

```python
query = "SELECT * FROM STUDENT_DEGREES;"
sql(query, engine=ENGINE)
```

```python
# EXERCISE 1: GET LIST OF NAMES OF ACTORS

query = "SELECT first_name, last_name FROM STUDENT_NAMES WHERE major = 'ACTOR' "
sql(query, engine = ENGINE)
```

```python
# EXERCISE 2: CREATE A TABLE WITH THE FOLLOWING SCHEMA:
# |degree|major|avg_grade|

query = '''SELECT 
    b.degree AS Degree, 
    a.major AS Major, 
    AVG(c.grade) AS Average_Grade 
FROM STUDENT_NAMES a 
INNER JOIN STUDENT_DEGREES b ON a.id = b.id 
INNER JOIN STUDENT_GRADES c on c.id = b.id

GROUP BY b.id
'''
sql(query, engine = ENGINE)
```

```python
# EXERCISE 3: CREATE A TABLE WITH THE FOLLOWING SCHEMA:
# full_name | major | avg_gpa | n_degrees|
# Note: full name = first_name last_name (e.g., John Smith)
```

```python
query = '''
SELECT CONCAT(a.first_name, " ", a.last_name) AS full_name, a.major, AVG(a.gpa) as avg_gpa, COUNT(b.degree) AS n_degrees
FROM STUDENT_NAMES a
LEFT JOIN STUDENT_DEGREES b ON a.id = b.id 
LEFT JOIN STUDENT_GRADES c on c.id = b.id

GROUP BY b.id
'''

sql(query, engine = ENGINE)
```
